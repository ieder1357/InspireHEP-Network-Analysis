```python
df.dtypes # and their data types
```

```python
# convert the creation_date column to datetime
df['creation_date'] = pd.to_datetime(df['creation_date'],errors='coerce')
```

```python
df.dtypes
```

```python
# Let's merge the authors and co-authors columns into one. 
# And get rid of columns we won't need for aesthetic reasons
df['authors'] = df[['authors', 'co-authors']].apply(lambda (x,y): x+y, axis=1)
del df['abstract']
del df['standardized_keywords']
del df['free_keywords']
del df['co-authors']
```

```python
# create citation and reference masks
citation_mask = (df['citations'].astype(str)!='[]')
reference_mask = (df['references'].astype(str)!='[]')
```

```python
# cut away papers with zero citations
df = df.loc[citation_mask]
df.shape[0]
```

```python
# and papers with 0 references
df = df.loc[reference_mask]
df.shape[0]
```

```python
# for some reason there are papers without authors. Let's get rid of them
author_mask=(df['authors'].astype(str)!='[]')
df = df.loc[author_mask]
df.shape[0]
```

```python
# get rid of papers whose titles contain certain keywords
title_mask = r'proceedings|withdrawn|Withdrawn|Proceedings|Thesis|thesis|Canceled|canceled|Cancelled|cancelled|Conference|conference'
df = df[~df.title.str.contains(title_mask)]
df.shape[0]
```

```python
# let's apply a time cut mask. i.e. only keep papers published in last 20 years
start_date = '06-01-1997'
end_date = '06-30-2017'
mask = ((df['creation_date'] > start_date) &  (df['creation_date'] < end_date))
df = df.loc[mask]
df.shape[0]
```

```python
df[:5] # quick peek
```

Ok, we are done cleaning the data. 
Let's write the relevant columns to lists now

```python
recids = df.recid.tolist()
authors = df.authors.tolist()
citations = df.citations.tolist()
references = df.citations.tolist()

len(references)
```

```python
# build network now
DG = nx.DiGraph() # directed graph where edges are (paper that cites, paper that's cited)
authors_in_recid = defaultdict(list) # key is recid, value is author list of recid
for i,recid in enumerate(recids):
    for reference in references[i]:
        DG.add_edge(recid,reference) # edge from recid to reference
    for citation in citations[i]:
        DG.add_edge(citation,recid) # edge from citation to recid
    authors_in_recid[recid] = authors[i]
```

```python
pr_by_recid = nx.pagerank(DG) # calculate the pagerank. Use default alpha=0.85
```

```python
# let's sort the entries into a list
pr_by_recid_sorted = sorted(pr_by_recid.items(),key=lambda x: x[1], reverse=True) 
```

```python
# Let's save the PageRank for future purposes
with open('data/recid_pr_last20years.dat','w') as f:
    for recid_i,pr_i in pr_by_recid_sorted:
        f.write(str(recid_i)+'\t'+str(pr_i)+'\n')
```

```python
# so in the future we can build the network from an input file
pr_by_recid_sorted =[]
pr_by_recid=defaultdict(float)
with open('data/recid_pr_last20years.dat') as f:
    for line in f:
        line = line.split()
        recid_i = int(line[0])
        pr_i = float(line[1])
        pr_by_recid_sorted.append((recid_i,pr_i))
        pr_by_recid[recid_i]=pr_i
```

Let's focus on authors now, and try to be more quantitative about citations vs PageRank-based impact

```python
# let's build the cumulative authors' PageRank-based measure
# we'll assign a paper's PR divided equally to its authors

pr_by_author = defaultdict(float)

for recid_i,authors_i in authors_in_recid.iteritems():
    for author_i in authors_i:
        pr_by_author[author_i] += pr_by_recid[recid_i]/len(authors_i)
pr_by_author_sorted = sorted(pr_by_author.items(),key=lambda x: x[1],reverse=True)
```

```python
# let's plot the top ranked authors

num_entries = 30
cumu_pr = [pr_i for _, pr_i in pr_by_author_sorted[:num_entries]]

x_ticks_labels = [author_i for author_i,_ in pr_by_author_sorted[:num_entries]]

x = range(num_entries)

plt.bar(x,cumu_pr)
plt.ylabel("Cumulative PageRank")
plt.xlabel("Author")
plt.title("Cumulative PageRank by Author from 1997-2017 (Author penalty = 1/N)")
plt.xticks([i+0.5 for i,_ in enumerate(cumu_pr)],x_ticks_labels,rotation='vertical')
plt.show()
```

# Are PageRank and Total Citation Count related?

Ok, is there a linear relationship between an author's total number of citations and their 
cumulative PR? Let's find out!

```python
# we need to build the num_cites_by_author dict and sorted list

num_cites_by_author = defaultdict(int) # key is author; value is total citation count

for i,_ in enumerate(recids):
    for author in authors[i]:
        num_cites_by_author[author] += len(citations[i])        
```

```python
# let's build arrays now for pr and for num_cites
num_cites = []
prs = []
for au,pr_i in pr_by_author_sorted:
    num_cites.append(num_cites_by_author[au])
    prs.append(pr_i)
```

```python
# Let's plot (some of) the data
plt.scatter(num_cites,prs)
plt.axis([10,200000,1e-10,0.002])
plt.yscale('log')
plt.xscale('log')
plt.xlabel("Number of Citations")
plt.ylabel("PageRank-based Ranking")
plt.show()
```

While there is definitely a correlation, there is a huuuge spread!
Let's fit the data to a straight line to better quantify this.
Note, when doing linear regression, one usually wants to split the dataset into a training sample to build and fit a model's parameters, and a test sample where we can try to predict the values with the model we built. 

But here, we already know from the above figure that a linear fit won't be effective at predicting the PR of a paper or author as a function of citations. 

So, let me just use linear regression to **find** a straight line. And I'll treat the **entire** dataset as my training sample because I'm not trying to predict some further datapoints. All I want to do for now is to find outliers w/r to some expectation under the hypothesis of there being a linear relationship between total citation count and PR.

```python
# convert arrays to numpy arrays, and reshape the independent variable
num_cites_np = np.array(num_cites).reshape(-1,1)
prs_np = np.array(prs)
```

```python
num_train = prs_np.shape[0] # usually we'd pick a subset of the data
x_train = num_cites_np[:num_train]
y_train = prs_np[:num_train]
x_test = num_cites_np[num_train:]
y_test = prs_np[num_train:]
```

```python
regr=linear_model.LinearRegression() 
```

```python
regr.fit(x_train,y_train)
```

```python
print regr.intercept_,regr.coef_ # coefficient alpha and beta
```

```python
y_predict = regr.predict(x_train) # usually you'd want to predict y for x_test as an input
```

```python
# let's see the data against a best-fit line
num_entries=prs_np.shape[0]
plt.scatter(x_train[:num_entries], y_train[:num_entries],  color='black')
plt.plot(x_train, y_predict,'.',color='blue')
plt.axis([50,100000,1e-8,0.0005])
#plt.ylim([1e-8,1e-4])
plt.yscale('log')
plt.xscale('log')
plt.xlabel("Number of Citations")
plt.ylabel("PageRank-based Ranking")
plt.show()
```

So, what would be really interesting is if we could find features or indicators that allow us to predict a high PR paper or author. But for now, we'll just examine some of the biggest outliers.

For this, we need the error bars on the estimated PR(num_citations) first. For now, since we are not trying to make predictions about our model on some test region, we will include the whole dataset (even the training region). So we are calculating the outliers w/r to the best-fit line

```python
# compute the standard deviation of the dependent variable PR(num_citations)
sigma = (prs_np - regr.predict(num_cites_np))**2
std_dev = math.sqrt(sigma.sum()/(prs_np.shape[0]-2)) # two degrees of freedom alpha and beta in fit.
print std_dev # standard deviation
```

```python
# compute residuals
num_of_std_devs = 5 # arbitrary. We want the biggest outliers. So let's pick "5 sigma"

residuals = prs_np - (regr.predict(num_cites_np) 
                      + num_of_std_devs*std_dev)
```

```python
res_index_sorted = residuals.argsort()[::-1] # get indices of outliers, sorted decreasingly
```

```python
# the biggest 10 outliers
print 'Name', 'Citations','PageRank','Residual'
for i in res_index_sorted[:10]:
    print pr_by_author_sorted[i][0],num_cites[i],prs[i],residuals[i]
```

Ok, so the top 10 outliers w/r to the hypothesis that "PR is linearly proportional to citations" are some of the really famous people we have seen before.
